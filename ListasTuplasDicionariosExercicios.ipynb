```python
'''
TAREFA 1
Escreva um programa que solicita que o
usuário digite 5 números inteiros e os coloca
em uma lista.
Quando terminar, imprima a lista.
'''
# Inicializa uma lista vazia para armazenar os números.
lista = []

# Solicita ao usuário que digite 5 números inteiros e os adiciona à lista.
for i in range(5):
  numero = int(input("Digite um número inteiro: "))
  lista.append(numero)

# Imprime a lista completa de números digitados.
print(lista)
```

```python
'''
TAREFA 2
Escreva um programa que aplica a função
módulo a todos os elementos e uma lista.
'''
# Lista de exemplo com números inteiros.
lista = [1, 2, 3, 4, 5]

# Percorre a lista e aplica a função abs() (módulo) a cada elemento.
# A função abs() retorna o valor absoluto de um número.
for i in range(len(lista)):
  lista[i] = abs(lista[i])

# Imprime a lista após aplicar a função módulo.
print(lista)
```

```python
'''
TAREFA 3
Escreva um programa que apaga todos os
elementos negativos de uma lista.
'''
# Lista de exemplo com números positivos e negativos.
lista = [1, -2, 3, -4, 5]

# Percorre a lista de trás para frente para evitar problemas de índice
# ao remover elementos. Se iterarmos para frente, a remoção de um elemento
# altera o índice dos elementos seguintes, podendo pular ou falhar.
for i in range(len(lista)-1, -1, -1):
  # Verifica se o elemento atual é negativo.
  if lista[i] < 0:
    # Remove o elemento negativo da lista.
    lista.pop(i)

# Imprime a lista após a remoção dos números negativos.
print(lista)
```

```python
'''
TAREFA 4
Faça um programa que solicita uma
quantidade indefinida de notas de uma
prova (deve parar quando um número
negativo for digitado).
Armazena em uma lista.
Após isso, o programa deve mostrar a média
das notas.
'''
# Inicializa uma lista vazia para armazenar as notas.
lista = []

# Loop para solicitar notas até que um número negativo seja digitado.
while True:
  nota = float(input("Digite uma nota: "))
  if nota < 0:
    break # Sai do loop se a nota for negativa.
  lista.append(nota) # Adiciona a nota à lista.

# Verifica se a lista de notas não está vazia para evitar ZeroDivisionError.
if len(lista) > 0:
  # Calcula a média das notas.
  media = sum(lista)/len(lista)
  print("A média das notas é: ", media)
else:
  print("Nenhuma nota válida foi digitada para calcular a média.")
```

```python
'''
TAREFA 5
Faça um programa que solicita uma
quantidade indefinida de números positivos
(deve parar quando um número negativo for
digitado).
Armazene-os em duas listas:
uma para os números pares;
outra para os números ímpares.
A seguir, mostre a porcentagem de pares e
ímpares digitados.
'''
# Inicializa listas vazias para números pares e ímpares.
pares = []
impares = []

# Loop para solicitar números até que um número negativo seja digitado.
while True:
  numero = int(input("Digite um número: "))
  if numero < 0:
    break # Sai do loop se o número for negativo.
  # Classifica o número como par ou ímpar e o adiciona à respectiva lista.
  if numero % 2 == 0:
    pares.append(numero)
  else:
    impares.append(numero)

# Calcula o total de números digitados (excluindo o negativo de parada).
total_numeros = len(pares) + len(impares)

# Verifica se algum número válido foi digitado para evitar ZeroDivisionError.
if total_numeros > 0:
  # Calcula a porcentagem de números pares.
  porcentagem_pares = (len(pares) / total_numeros) * 100
  # Calcula a porcentagem de números ímpares.
  porcentagem_impares = (len(impares) / total_numeros) * 100

  print("Porcentagem de pares: ", porcentagem_pares)
  print("Porcentagem de ímpares: ", porcentagem_impares)
else:
  print("Nenhum número positivo foi digitado para calcular as porcentagens.")
```

```python
'''
TAREFA 6
Faça um programa que solicita uma
quantidade indefinida de números positivos
(deve parar quando um número negativo for
digitado).
Armazene todos os números digitados em
uma lista, sem repetição.
'''
# Inicializa uma lista vazia para armazenar os números únicos.
lista = []

# Loop contínuo para solicitar números ao usuário.
while True:
  numero = int(input("Digite um número: "))
  # Se o número digitado for negativo, o loop é interrompido.
  if numero < 0:
    break
  # Verifica se o número já existe na lista. Se não existir, ele é adicionado.
  # Isso garante que a lista não terá números repetidos.
  if numero not in lista:
    lista.append(numero)

# Imprime a lista final contendo apenas números positivos e sem repetições.
print(lista)
```

```python
'''
TAREFA 7
Escreva um programa que solicita que o
usuário digite a altura e o peso de 5 pessoas.
Verifique se pelo menos duas das pessoas
tem a mesma altura e mesmo peso.
'''
# Inicializa listas vazias para armazenar as alturas e pesos.
lista_altura = []
lista_peso = []

# Loop para solicitar a altura e o peso de 5 pessoas.
for i in range(5):
  altura = float(input(f"Digite a altura da pessoa {i+1}: "))
  peso = float(input(f"Digite o peso da pessoa {i+1}: "))
  lista_altura.append(altura)
  lista_peso.append(peso)

# Variável de controle para verificar se há pessoas com altura e peso iguais.
pessoas_iguais_encontradas = False

# Compara cada pessoa com as pessoas seguintes para encontrar duplicatas.
for i in range(len(lista_altura)):
  # Começa a comparação a partir do próximo elemento (i+1) para evitar
  # comparar uma pessoa com ela mesma e evitar comparações redundantes.
  for j in range(i+1, len(lista_altura)):
    # Verifica se a altura da pessoa 'i' é igual à altura da pessoa 'j'
    # E se o peso da pessoa 'i' é igual ao peso da pessoa 'j'.
    if lista_altura[i] == lista_altura[j] and lista_peso[i] == lista_peso[j]:
      print("Pelo menos duas pessoas têm a mesma altura e mesmo peso.")
      pessoas_iguais_encontradas = True
      break # Sai do loop interno assim que uma duplicata é encontrada.
  if pessoas_iguais_encontradas:
    break # Sai do loop externo se uma duplicata já foi encontrada.

# Se nenhum par de pessoas com altura e peso iguais for encontrado após todas as comparações.
if not pessoas_iguais_encontradas:
    print("Não foram encontradas duas pessoas com a mesma altura e mesmo peso.")
```

```python
'''
TAREFA 8
Faça um programa que solicita que o
usuário digite o nome e a idade de diversas
pessoas (o programa deve parar quando
um nome vazio for informado).
Retorne o nome da pessoa mais velha.
'''
# Inicializa listas vazias para armazenar nomes e idades.
lista_nome = []
lista_idade = []

# Loop contínuo para solicitar nome e idade das pessoas.
while True:
  nome = input("Digite o nome (deixe em branco para parar): ")
  # Se o nome digitado for vazio, o loop é interrompido.
  if nome == "":
    break
  idade = int(input(f"Digite a idade de {nome}: "))

  # Adiciona o nome e a idade às respectivas listas.
  lista_nome.append(nome)
  lista_idade.append(idade)

# Verifica se alguma pessoa foi digitada antes de tentar encontrar a mais velha.
if lista_idade:
  # Encontra a maior idade na lista de idades.
  maior_idade = max(lista_idade)
  # Encontra o índice da maior idade. Se houver idades repetidas, retorna o primeiro índice.
  indice_maior_idade = lista_idade.index(maior_idade)
  # Usa o índice para obter o nome da pessoa mais velha da lista de nomes.
  nome_maior_idade = lista_nome[indice_maior_idade]

  # Imprime o nome da pessoa mais velha.
  print("A pessoa mais velha é: ", nome_maior_idade)
else:
  print("Nenhum nome e idade foram digitados.")
```

```python
'''
TAREFA 1
Dado o dicionário ao lado,
= {
"Alice": 8.5,
"Bruno": 7.8,
"Carla": 9.2,
"Diego": 6.9,
"Eva": 8.0,
"Fernando": 7.5,
"Gabriela": 9.0,
"Henrique": 6.7,
"Isabela": 8.3,
"João": 7.2
}
imprima os alunos seguintes dados:

Quantos alunos existem?
Alguém tirou nota zero?
Qual foi a menor nota?
Qual foi a maior nota?
'''

notas_alunos = {
    "Alice": 8.5,
    "Bruno": 7.8,
    "Carla": 9.2,
    "Diego": 6.9,
    "Eva": 8.0,
    "Fernando": 7.5,
    "Gabriela": 9.0,
    "Henrique": 6.7,
    "Isabela": 8.3,
    "João": 7.2
}

# Quantos alunos existem?
num_alunos = len(notas_alunos)
print(f"Número de alunos: {num_alunos}")

# Alguém tirou nota zero?
notas = list(notas_alunos.values())
alguem_tirou_zero = 0.0 in notas
print(f"Alguém tirou nota zero? {alguem_tirou_zero}")

# Qual foi a menor nota?
menor_nota = min(notas)
print(f"Menor nota: {menor_nota}")

# Qual foi a maior nota?
maior_nota = max(notas)
print(f"Maior nota: {maior_nota}")
```

```python
'''
TAREFA 2
Dado o dicionário ao lado:

animes = { "Naruto": 220,
"Jujutsu Kaisen": 47, "Dragon Ball Z": 291, "Death Note": 37,
"Fullmetal Alchemist": 64,
"Evangelion": 26,
"Berserk": 25,
"Code Geass": 50, "Akame ga Kill!": 24, "Elfen Lied": 13
}

Imprima o total de episódios.
Substitua a quantidade de episódios pelo percentual (ep_anime/ep_total).
Imprima nome e percentual de cada anime, em ordem decrescente de percentual.
'''

animes = {
    "Naruto": 220,
    "Jujutsu Kaisen": 47,
    "Dragon Ball Z": 291,
    "Death Note": 37,
    "Fullmetal Alchemist": 64,
    "Evangelion": 26,
    "Berserk": 25,
    "Code Geass": 50,
    "Akame ga Kill!": 24,
    "Elfen Lied": 13
}

# Imprima o total de episódios.
total_episodios = sum(animes.values())
print(f"Total de episódios: {total_episodios}")

# Substitua a quantidade de episódios pelo percentual (ep_anime/ep_total).
# Crie um novo dicionário para armazenar os percentuais
animes_percentual = {}
for nome_anime, num_episodios in animes.items():
    percentual = (num_episodios / total_episodios) * 100
    animes_percentual[nome_anime] = percentual

# Imprima nome e percentual de cada anime, em ordem decrescente de percentual.
animes_ordenados = sorted(animes_percentual.items(), key=lambda item: item[1], reverse=True)

print("\nAnimes por percentual de episódios (decrescente):")
for nome_anime, percentual in animes_ordenados:
    print(f"{nome_anime}: {percentual:.2f}%")
```

```python
'''
TAREFA 3
Dado o dicionário ao lado:

bandas = {
"Gangrena Gasosa": [
{"álbum": "Gente Ruim Só Manda Lembrança Pra Quem Não Presta", "ano":
2019},
{"álbum": "Se Deus É 10, Satanás É 666", "ano": 1999},
{"álbum": "Welcome to the Macumba", "ano": 1993}
],
"Rogério Skylab": [
{"álbum": "Nas Portas do Cu", "ano": 2023},
{"álbum": "Crítica da Faculdade do Cu", "ano": 2021},
{"álbum": "Cosmos", "ano": 2019}
],
"Garotos Podres": [
{"álbum": "Canções para Ninar", "ano": 2003},
{"álbum": "Com a Corda Toda", "ano": 1997},
{"álbum": "Rock de Subúrbio", "ano": 1995}
],
"Massacration": [
{"álbum": "Live Metal Espancation", "ano": 2017},
{"álbum": "Good Blood Headbanguers", "ano": 2009},
{"álbum": "Gates of Metal Fried Chicken of Death", "ano": 2005}
],
"Raimundos": [
{"álbum": "Cantigas de Roda", "ano": 2014},
{"álbum": "Roda Viva", "ano": 2006},
{"álbum": "Éramos 4", "ano": 2001}
]
}

Imprima o álbum mais antigo.
Imprima a banda que tem o maior intervalo entre os álbuns.
Imprima todos os álbuns do mais novo para o mais antigo, dizendo também o nome da banda.
'''

bandas = {
    "Gangrena Gasosa": [
        {"álbum": "Gente Ruim Só Manda Lembrança Pra Quem Não Presta", "ano": 2019},
        {"álbum": "Se Deus É 10, Satanás É 666", "ano": 1999},
        {"álbum": "Welcome to the Macumba", "ano": 1993}
    ],
    "Rogério Skylab": [
        {"álbum": "Nas Portas do Cu", "ano": 2023},
        {"álbum": "Crítica da Faculdade do Cu", "ano": 2021},
        {"álbum": "Cosmos", "ano": 2019}
    ],
    "Garotos Podres": [
        {"álbum": "Canções para Ninar", "ano": 2003},
        {"álbum": "Com a Corda Toda", "ano": 1997},
        {"álbum": "Rock de Subúrbio", "ano": 1995}
    ],
    "Massacration": [
        {"álbum": "Live Metal Espancation", "ano": 2017},
        {"álbum": "Good Blood Headbanguers", "ano": 2009},
        {"álbum": "Gates of Metal Fried Chicken of Death", "ano": 2005}
    ],
    "Raimundos": [
        {"álbum": "Cantigas de Roda", "ano": 2014},
        {"álbum": "Roda Viva", "ano": 2006},
        {"álbum": "Éramos 4", "ano": 2001}
    ]
}

# Encontre e imprima o álbum mais antigo.
ano_album_mais_antigo = float('inf')
nome_album_mais_antigo = ""
nome_banda_album_mais_antigo = ""

for nome_banda, albuns_banda in bandas.items():
    for album in albuns_banda:
        if album['ano'] < ano_album_mais_antigo:
            ano_album_mais_antigo = album['ano']
            nome_album_mais_antigo = album['álbum']
            nome_banda_album_mais_antigo = nome_banda

print(f"\nÁlbum mais antigo: '{nome_album_mais_antigo}' da banda {nome_banda_album_mais_antigo} ({ano_album_mais_antigo})")

# Encontre e imprima a banda que tem o maior intervalo entre os álbuns.
maior_intervalo = -1
banda_com_maior_intervalo = ""

for nome_banda, albuns_banda in bandas.items():
    anos_lancamento = [album['ano'] for album in albuns_banda]
    if len(anos_lancamento) > 1:
        intervalo_atual = max(anos_lancamento) - min(anos_lancamento)
        if intervalo_atual > maior_intervalo:
            maior_intervalo = intervalo_atual
            banda_com_maior_intervalo = nome_banda

print(f"\nBanda com o maior intervalo entre os álbuns: {banda_com_maior_intervalo} ({maior_intervalo} anos)")

# Imprima todos os álbuns do mais novo para o mais antigo, dizendo também o nome da banda.
todos_albuns_planos = []
for nome_banda, albuns_banda in bandas.items():
    for album in albuns_banda:
        todos_albuns_planos.append({"banda": nome_banda, "álbum": album['álbum'], "ano": album['ano']})

albuns_ordenados_por_ano = sorted(todos_albuns_planos, key=lambda x: x['ano'], reverse=True)

print("\nTodos os álbuns (do mais novo para o mais antigo):")
for dados_album in albuns_ordenados_por_ano:
    print(f"'{dados_album['álbum']}' - {dados_album['banda']} ({dados_album['ano']})")
```
