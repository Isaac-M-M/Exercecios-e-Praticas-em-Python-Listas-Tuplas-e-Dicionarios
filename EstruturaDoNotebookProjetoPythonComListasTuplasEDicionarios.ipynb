# **1. üõí E-commerce**

```python
# Cat√°logo de produtos: lista de tuplas (nome, pre√ßo, categoria)
catalogo = [
  ("Smartphone", 1500, "Eletr√¥nicos"),
  ("T√™nis", 250, "Moda"),
  ("Notebook", 3500, "Eletr√¥nicos"),
  ("Livro", 40, "Educa√ß√£o"),
  ("Fone de ouvido", 200, "Eletr√¥nicos")
]

# Filtrar produtos da categoria "Eletr√¥nicos" usando list comprehension
eletronicos = [produto for produto in catalogo if produto [2] == "Eletr√¥nicos"]
# Verificar se a lista de eletr√¥nicos tem 3 elementos (Smartphone, Notebook, Fone de ouvido)
assert len(eletronicos) == 3
# Imprimir a lista de produtos eletr√¥nicos
print("Produtos eletr√¥nicos:", eletronicos)

# Carrinho de compras: dicion√°rio onde a chave √© o nome do produto e o valor √© a quantidade
carrinho = {}
# Adicionar Notebook ao carrinho (1 unidade)
carrinho ["Notebook"] = 1
# Adicionar Fone de ouvido ao carrinho (2 unidades)
carrinho ["Fone de ouvido"] = 2
# Remover Notebook do carrinho
del carrinho ["Notebook"]

# Verificar se Notebook n√£o est√° mais no carrinho
assert "Notebook" not in carrinho
# Verificar se a quantidade de Fone de ouvido no carrinho √© 2
assert carrinho ["Fone de ouvido"] == 2
# Imprimir o conte√∫do do carrinho
print("Carrinho:", carrinho)

# Lista de pre√ßos de alguns produtos
precos = [1500, 200, 3500]
# Calcular pre√ßos com 10% de desconto usando list comprehension
precos_com_desconto = [p *0.9 for p in precos]
# Calcular o total dos pre√ßos com desconto
total = sum(precos_com_desconto)

# Verificar se o total calculado √© igual √† soma dos pre√ßos originais com 10% de desconto
assert round(total, 2) == round(1500*0.9 + 200*0.9+ 3500*0.9, 2)
# Imprimir o total com desconto formatado como moeda (R$)
print(f"Total com desconto: R$ {total:.2f}")

# Estoque de produtos por categoria: dicion√°rio onde a chave √© a categoria e o valor √© uma lista de produtos
estoque = {
"Eletr√¥nicos": ["Smartphone", "Notebook", "Fone de ouvido"], "Moda": ["T√™nis", "Camisa"],
"Educa√ß√£o": ["Livro"]
}
# Encontrar a categoria com o maior n√∫mero de produtos usando a fun√ß√£o max e uma fun√ß√£o lambda
categoria_maior = max(estoque, key=lambda c: len(estoque[c]))
# Verificar se a categoria com mais produtos √© "Eletr√¥nicos"
assert categoria_maior == "Eletr√¥nicos"
# Imprimir a categoria com mais produtos
print("Categoria com mais produtos:", categoria_maior)
```

# 2. üïπ Jogos

```python
# Ranking de jogadores: lista de tuplas (nome, pontua√ß√£o)
ranking = [("Ana", 120), ("Bruno", 200), ("Carlos", 180)]
# Ordenar o ranking por pontua√ß√£o em ordem decrescente usando uma fun√ß√£o lambda
ranking.sort(key=lambda x: x[1], reverse=True)

# Verificar se o primeiro jogador no ranking √© "Bruno"
assert ranking [0] [0] == "Bruno"
# Imprimir o top 3 jogadores
print("Top 3 jogadores:", ranking[:3])

# Invent√°rio de itens: dicion√°rio onde a chave √© o nome do item e o valor √© uma tupla (quantidade, raridade)
inventario = {
  "Espada": (1, "raro"),
  "Po√ß√£o": (5, "comum"),
  "Armadura": (2, "√©pico")
}
# Filtrar itens com raridade "raro" ou "√©pico" usando list comprehension
raros = [item for item, (_, raridade) in inventario.items() if raridade in ["raro", "√©pico"]]
# Verificar se "Espada" e "Armadura" est√£o na lista de itens raros
assert "Espada" in raros and "Armadura" in raros
# Imprimir a lista de itens raros
print("Itens raros:", raros)

# Partidas de jogo: lista de tuplas (jogador1, jogador2, vencedor)
partidas = [
  ("Ana", "Bruno", "Ana"),
  ("Carlos", "Ana", "Carlos"),
  ("Bruno", "Carlos", "Carlos")
]
# Contar o n√∫mero de vit√≥rias de cada jogador usando um dicion√°rio
vitorias = {}
for _, _, vencedor in partidas:
  vitorias [vencedor] = vitorias.get(vencedor, 0) + 1

# Verificar se o n√∫mero de vit√≥rias de "Carlos" √© 2
assert vitorias ["Carlos"] == 2
# Imprimir o dicion√°rio de vit√≥rias
print("Vit√≥rias:", vitorias)

# Configura√ß√µes do jogo: dicion√°rio com as configura√ß√µes atuais
config = {
  "volume": 70,
  "dificuldade": "m√©dio",
  "idioma": "portugu√™s"
}

# Alterar o volume para 50
config["volume"] = 50
# Alterar o idioma para "ingl√™s"
config["idioma"] = "ingl√™s"

# Verificar se o volume foi alterado para 50
assert config["volume"] == 50
# Imprimir as configura√ß√µes atuais
print("Configura√ß√µes:", config)
```

# 3. üéµ Playlist

```python
# Playlist de m√∫sicas: lista de dicion√°rios com informa√ß√µes sobre cada m√∫sica
playlist = [
  {"nome": "M√∫sica A", "artista": "Artista 1", "duracao": 2.5},
  {"nome": "M√∫sica B", "artista": "Artista 2", "duracao": 3.2},
  {"nome": "M√∫sica C", "artista": "Artista 1", "duracao": 4.0}
]

# Filtrar m√∫sicas com dura√ß√£o maior que 3 minutos usando list comprehension
longas = [m["nome"] for m in playlist if m["duracao"] > 3]
# Verificar se "M√∫sica B" e "M√∫sica C" est√£o na lista de m√∫sicas longas
assert "M√∫sica B" in longas and "M√∫sica C" in longas
# Imprimir a lista de m√∫sicas longas
print("M√∫sicas com mais de 3 minutos:", longas)

# M√∫sicas favoritas: dicion√°rio onde a chave √© o nome da m√∫sica e o valor indica se √© favorita (True/False)
favoritas = {
  "M√∫sica A": False,
  "M√∫sica B": True,
  "M√∫sica C": True
}

# Criar uma lista das m√∫sicas favoritas usando list comprehension
favoritas_listadas = [nome for nome, fav in favoritas.items() if fav]
# Verificar se a lista de favoritas tem 2 elementos
assert len(favoritas_listadas) == 2
# Imprimir a lista de m√∫sicas favoritas
print("Favoritas:", favoritas_listadas)

# Lista de m√∫sicas com nome e dura√ß√£o
musicas = [("M√∫sica A", 2.5), ("M√∫sica B", 3.2), ("M√∫sica C", 4.0)]
# Calcular a dura√ß√£o total das m√∫sicas
total = sum([dur for _, dur in musicas])
# Calcular a dura√ß√£o m√©dia das m√∫sicas
media = total / len(musicas)

# Verificar se a m√©dia calculada √© igual √† soma total dividida por 3
assert round(media, 2) == round(total / 3, 2)
# Imprimir a dura√ß√£o total e a m√©dia formatadas
print(f"Total: {total:.2f} min M√©dia: {media:.2f} min")

# M√∫sicas por artista: dicion√°rio onde a chave √© o artista e o valor √© uma lista de suas m√∫sicas
por_artista = {
  "Artista 1": ["M√∫sica A", "M√∫sica C"],
  "Artista 2": ["M√∫sica B"]
}
# Encontrar o artista com mais m√∫sicas usando a fun√ß√£o max e uma fun√ß√£o lambda
artista_top = max(por_artista, key=lambda a: len(por_artista[a]))
# Verificar se o artista com mais m√∫sicas √© "Artista 1"
assert artista_top == "Artista 1"
# Imprimir o artista com mais m√∫sicas
print("Artista com mais m√∫sicas:", artista_top)
```
