```python
'''
15. Refazer o programa 14 realizando as quatro operações aritméticas básicas
'''
# Solicita o primeiro número real ao usuário e o converte para float.
real1 = float(input('Digite o primeiro número real: '))
# Solicita o segundo número real ao usuário e o converte para float.
real2 = float(input('Digite o segundo número real: '))

# Realiza e imprime o produto.
print('O produto dos numeros', real1, 'e', real2, 'corresponde a', real1 * real2)
# Realiza e imprime a soma.
print('A soma dos numeros', real1, 'e', real2, 'corresponde a', real1 + real2)
# Realiza e imprime a subtração.
print('A subtração dos numeros', real1, 'e', real2, 'corresponde a', real1 - real2)
# Realiza e imprime a divisão.
# Adiciona uma verificação para evitar divisão por zero.
if real2 != 0:
    print('A divisão dos numeros', real1, 'e', real2, 'corresponde a', real1 / real2)
else:
    print('Não é possível dividir por zero.')
```

```python
'''
16. Escreva um programa que solicite o valor fixo do salário de um vendedor, o total vendido
no mês e o percentual de comissão do vendedor. Ao final apresentar o salário bruto.
'''
# Solicita o valor fixo do salário e converte para float.
salario_fixo = float(input('Digite o valor fixo do salário de um vendedor: '))
# Solicita o total vendido no mês e converte para float.
total_vendido = float(input('Digite o total vendido no mês: '))
# Solicita o percentual de comissão e converte para float.
percentual_comissao = float(input('Digite o percentual de comissão do vendedor: '))

# Calcula o valor da comissão.
valor_comissao = total_vendido * (percentual_comissao / 100)
# Calcula o salário bruto.
salario_bruto = salario_fixo + valor_comissao
# Imprime o salário bruto.
print(f'O salário bruto do vendedor é: R$ {salario_bruto:.2f}')
```
