```python
'''
13. Elabore um algoritmo que leia um número de entrada que indicará a quantidade de
registros a serem lidos (N). Em seguida algoritmo deve solicitar o sexo (M/F) e idade de N
pessoas e ao final apresentar a média de idade de ambos os gêneros catalogados.
'''
# Solicita ao usuário a quantidade de registros (N) a serem lidos
n = int(input('Digite a quantidade de registros: '))
# Inicializa variáveis para somar as idades de homens e mulheres
soma_m = 0
soma_f = 0
# Inicializa contadores para a quantidade de homens e mulheres
quantidade_m = 0
quantidade_f = 0

# Loop para iterar N vezes, lendo os registros de cada pessoa
for i in range(n):
  # Solicita o sexo (M/F) da pessoa
  sexo = input('Digite o sexo (M/F): ')
  # Solicita a idade da pessoa
  idade = int(input('Digite a idade: '))
  # Verifica o sexo informado e atualiza as somas e contadores correspondentes
  if sexo == 'M':
    soma_m = soma_m + idade
    quantidade_m = quantidade_m + 1
  else:
    soma_f = soma_f + idade
    quantidade_f = quantidade_f + 1

# Calcula e imprime a média de idade dos homens (evitando divisão por zero)
if quantidade_m > 0:
  print('A média de idade dos homens é: ', soma_m/quantidade_m)
else:
  print('Não foram informadas idades para homens.')

# Calcula e imprime a média de idade das mulheres (evitando divisão por zero)
if quantidade_f > 0:
  print('A média de idade das mulheres é: ', soma_f/quantidade_f)
else:
  print('Não foram informadas idades para mulheres.')
```

```python
'''
14. Elabore um algoritmo que solicite ao usuário 10 números reais e ao final apresente o
maior e o menor deles.
'''

# Informa ao usuário que ele deverá digitar 10 números reais
print('Digite 10 números reais')
# Inicializa 'maiorN' e 'menorN' com 0 (este é um ponto de atenção, pois se todos os números forem negativos, o maior estará incorreto)
# Uma abordagem melhor seria inicializar 'maiorN' com o primeiro número lido e 'menorN' com o primeiro número lido também.
maiorN = 0
menorN = 0

# Loop para iterar 10 vezes
for i in range(10):
  # Solicita um número real ao usuário
  n = float(input('Número:'))
  # Comentários incompletos/lógica inicial para maior e menor precisa ser revisada
  # A lógica atual apenas verifica se o número é maior que 0 ou menor que 0
  # e não compara com os números anteriores para encontrar o maior e o menor de *todos* os números.
  # Para uma solução correta, seria necessário um 'if' para o primeiro número
  # ou inicializar maiorN/menorN com um valor muito baixo/alto ou com o primeiro input.
  if i == 0: # Para o primeiro número, inicializa maiorN e menorN
      maiorN = n
      menorN = n
  else:
      if n > maiorN:
          maiorN = n
      if n < menorN:
          menorN = n

# Imprime o maior e o menor número encontrados
print('O maior número é:', maiorN, 'e o menor número é:', menorN)
```

```python
'''
15. Elabore um algoritmo que solicite N números reais e quando o usuário informar o valor
nulo 0 (zero) o programa ordene e mostre todos os números informados de forma crescente.
'''
# Cria uma lista vazia para armazenar os números digitados pelo usuário
numeros = []

# Loop infinito que continuará solicitando números até que o usuário digite 0
while True:
  # Solicita ao usuário um número real e o converte para float
  n = float(input('Digite um número real (ou 0 para parar): '))
  # Verifica se o número digitado é 0
  if n == 0:
    break  # Se for 0, interrompe o loop (o usuário quer parar de inserir números)
  numeros.append(n)  # Se não for 0, adiciona o número à lista 'numeros'

# Ordena a lista de números em ordem crescente. O método .sort() modifica a lista in-place.
numeros.sort()

# Imprime um cabeçalho indicando que os números serão mostrados em ordem crescente
print('Números informados em ordem crescente:')
# Loop para percorrer a lista 'numeros' (já ordenada) e imprimir cada número
for numero in numeros:
  print(numero)
```

```python
'''
16. Escreva um programa que vá solicitando as idades dos alunos da sala até que todos
sejam informados (perguntar ao usuário se deseja informar a idade do próximo aluno). Ao
final apresentar a idade do mais novo, a idade do mais velho, Quantos alunos têm mais de
18 anos, quantos alunos têm até 18 anos, a média aritmética e a mediana.
'''

# Inicializa uma lista vazia para armazenar as idades dos alunos
idades = []

# Loop principal para solicitar as idades dos alunos
while True:
  # Solicita a idade do aluno ao usuário e converte para inteiro
  idade = int(input('Digite a idade do aluno: '))
  # Adiciona a idade à lista
  idades.append(idade)
  # Pergunta ao usuário se ele deseja informar a idade do próximo aluno
  continuar = input('Deseja informar a idade do próximo aluno? (sim/não): ').lower()
  # Se a resposta não for 'sim', interrompe o loop
  if continuar != 'sim':
    break

import numpy as np

# Calcula a idade do aluno mais novo da lista
mais_novo = min(idades)
# Calcula a idade do aluno mais velho da lista
mais_velho = max(idades)

# Inicializa contadores para alunos com mais de 18 anos e até 18 anos
maiores_de_18 = 0
ate_18 = 0
# Itera sobre cada idade na lista 'idades'
for idade in idades:
  # Verifica se a idade é maior que 18
  if idade > 18:
    # Incrementa o contador de maiores de 18
    maiores_de_18 += 1
  else:
    # Incrementa o contador de até 18 anos
    ate_18 += 1

# Calcula a média aritmética das idades (soma de todas as idades dividida pelo número total de idades)
media_idades = sum(idades) / len(idades)
# Calcula a mediana das idades usando a função median() da biblioteca numpy
mediana_idades = np.median(idades)

# Apresenta todos os resultados solicitados
print(f'Idade do mais novo: {mais_novo} anos')
print(f'Idade do mais velho: {mais_velho} anos')
print(f'Alunos com mais de 18 anos: {maiores_de_18}')
print(f'Alunos com até 18 anos: {ate_18}')
print(f'Média aritmética das idades: {media_idades:.2f} anos')
print(f'Mediana das idades: {mediana_idades} anos')
```
