```python
'''8. Elabore um algoritmo que leia um número, e se ele for maior do que 20, imprimir a metade
desse número, caso contrário, imprimir o dobro do número'''

# Solicita ao usuário que digite um número inteiro e armazena-o na variável 'n1'
n1 = input('Escreva um número inteiro: ')

# Converte a entrada para float e verifica se o número é maior que 20
if float(n1) > 20:
  # Se for maior que 20, calcula e imprime a metade do número
  print('A metade de ', n1, ' é: ', float(n1) / 2)
# Se a condição acima não for verdadeira, executa este bloco
elif float(n1):
  # Caso contrário (o número é menor ou igual a 20), calcula e imprime o dobro do número
  print('O dobro de ', n1, ' é: ', int(n1) * 2)
```

```python
'''9. Elabore um algoritmo que leia dois números inteiros e realize a adição; caso o resultado
seja maior que 10, imprima o quadrado do resultado, caso contrário, imprima a metade dele'''

# Solicita ao usuário o primeiro número inteiro e armazena-o em 'n1'
n1 = input('Escreva um número inteiro: ')
# Solicita ao usuário o segundo número inteiro e armazena-o em 'n2'
n2 = input('Escreva outro número inteiro: ')

# Verifica se a soma dos dois números (convertidos para float) é menor que 10
if float(n1) + float(n2) < 10:
  # Se a soma for menor que 10, calcula o quadrado da soma e o imprime
  print((float(n1) + float(n2)) * (float(n1) + float(n2)) )
# Se a soma não for menor que 10, verifica se ela é maior ou igual a 10
elif float(n1) + float(n2) >= 10:
  # Se a soma for maior ou igual a 10, calcula a metade da soma e a imprime
  print((float(n1) + float(n2)) / 2)
```

```python
'''10. O sistema de avaliação de determinada disciplina é composto por três provas. A primeira
prova tem peso 2, a segunda tem peso 3 e a terceira tem peso 5. Considerando que a
média para aprovação é 6.0, faça um algoritmo para calcular a média final de um aluno
desta disciplina e dizer se o aluno foi aprovado ou não'''

# Solicita ao usuário as notas das três provas e as armazena em 'n1', 'n2' e 'n3'
n1 = input('Escreva a nota da primeira prova: ')
n2 = input('Escreva a nota da segunda prova: ')
n3 = input('Escreva a nota da terceira prova: ')

# Calcula a média ponderada das notas, dividindo pela soma dos pesos (2+3+5=10)
media = (float(n1) * 2 + float(n2) * 3 + float(n3) * 5) / 10

# Verifica se a média é maior ou igual a 6 (média para aprovação)
if media >= 6:
  # Se a média for maior ou igual a 6, imprime 'Aprovado'
  print('Aprovado')
# Se a média não for maior ou igual a 6, executa este bloco
elif media < 6:
  # Imprime 'Reprovado'
  print('Reprovado')
```
